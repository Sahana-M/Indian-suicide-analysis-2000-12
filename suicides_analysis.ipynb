```python
import pandas as pd

df = pd.read_csv("suicides.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
dfcopy = df.copy()
df = df[df.State != 'Total (All India)']
df = df[df.State != 'Total (States)']
df = df[df.State != 'Total (Uts)']
print("before : ", len(dfcopy), " after : ", len(df))
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot of gender


female = df.groupby('Gender')['Total'].sum()['Female']
male = df.groupby('Gender')['Total'].sum()['Male']

index = ['Male', 'Female']
values = [male, female]

new = pd.DataFrame({'gender':index, 'val':values})
ax = new.plot.bar(x='gender', y='val', rot=0, title = "Gender wise suicides")

print("Male , Female : ",values )
```

```python
#plot of Years vs Suicides

l_values = list()
l_index = list()
np.asarray(l_values)
np.asarray(l_index)
l_values = df.groupby('Year')['Total'].sum().unique()
l_index = df.Year.unique()

new = pd.DataFrame({'years':l_index, 'suicides':l_values})
ax = new.plot.bar(x='years', y='suicides', rot=0, title = "Suicides Year wise")
```

```python
#plot of States vs Suicides


l_values = list()
l_index = list()
np.asarray(l_values)
np.asarray(l_index)

l_values = df.groupby('State')['Total'].sum()
l_index = df.State.unique()


new = pd.DataFrame({'States':l_index, 'suicides':l_values})
ax = new.plot.bar(x='States', y='suicides', rot=45, title = "Suicides State wise", figsize = (25,15))
```
